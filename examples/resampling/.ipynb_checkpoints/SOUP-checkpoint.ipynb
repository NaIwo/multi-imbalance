Unzip datasets and prepare data:

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.decomposition import PCA
from multi_imbalance.datasets import load_datasets
import multi_imbalance.resampling

from multi_imbalance.utils import *

%matplotlib inline
# sns.set_style('darkgrid')


ecoli_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/glass/glass.data'
df = pd.read_csv(ecoli_url, delimiter=',', header=None,
                 names=['name', '1', '2', '3', '4', '5', '6', '7', 'class'])
X, y = df.iloc[:, 1:8].to_numpy(), df['class'].to_numpy()
print(X[:5])
print(y[:5])
type(X[0][0])
```

```python
d = load_datasets()['ecoli']
X2, y2 = d.data, d.target
print(X2[:5])
print(y2[:5])
type(X2[0][0])
```

Resample data using SOUP algorithm

```python
clf = multi_imbalance.resampling.SOUP()
resampled_X, resampled_y = clf.fit_transform(X, y)
```

Compare results by plotting data in 2 dimensions

```python
pca = PCA(n_components=2)
pca.fit(X)

fig, axs = plt.subplots(ncols=2, nrows=2)
fig.set_size_inches( 16, 10)
axs = axs.flatten()

sns.countplot(y, ax=axs[0])
X = pca.transform(X)
df = construct_flat_2pc_df(X, y)
sns.scatterplot(x='x1', y='x2', hue='y', data=df, alpha=0.7, ax=axs[1], legend='brief')

sns.countplot(resampled_y, ax=axs[2])
resampled_X = pca.transform(resampled_X)
df = construct_flat_2pc_df(resampled_X, resampled_y)
sns.scatterplot(x='x1', y='x2', hue='y', data=df, alpha=0.7, ax=axs[3], legend='brief')
```
